#  EDA is exploratory data analysis and it is basically the analysis of data
## EDA has three steps to follow. These are not manadatory but they help you explore your data.

- Data profiling
- Statistical based analysis
- Graphical based analysis

## Data profiling
- How many rows and columns does my data have
- Are there missing values, duplicate values and outliers.
- Check for the data types
- The size of my data (RAM)
- Check whether my data is numerical, categorical or date
- Data cleaning

##  Statistical based analysis

- Variance, Standard deviation
- Mean, Median, Mode
- Skewness
- Range, Percentile
- Correlation and covariance
- Stat tests (T, f, Z, chisquare)

## Graphical based analysis

- Boxplot
- Scatter plot
- KDE 
- Count plot
- Histogram
- Heatmap

# Bivariate Analysis

```python
#Which Category has largest number of installation. This a bivariate analysis
df_cat_installs = df.groupby(["Category"])["Installs"].sum().sort_values(ascending=False).reset_index()
```

```python
df_cat_installs10= df_cat_installs.head(10)
```

```python
sns.barplot(x = "Installs" , y = "Category",data = df_cat_installs10)
plt.title('Top 10 Categories by Installs')
plt.xticks(rotation=90)
plt.show()
```

```python
# Beautified version

plt.figure(figsize = (14,10))
sns.set_context("talk")
sns.set_style("darkgrid")
ax = sns.barplot(x = 'Installs' , y = 'Category' , data =df_cat_installs10) 
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular Categories in Play Store", size = 20)
```

# Multivariate Analysis 

```python
# What are the Top 5 most installed Apps in Each popular Category
df1 = df.groupby(["Category","App"])["Installs"].sum().reset_index()
df1 = df1.sort_values("Installs", ascending = False)
```

```python
df1[df1.Category == "FAMILY"][:5]
```

```python
apps = ['GAME', 'COMMUNICATION', 'PHOTOGRAPHY', 'SOCIAL' ]
sns.set_context("poster")
sns.set_style("darkgrid")
plt.figure(figsize=(40,30))
for i,app in enumerate(apps):
    df2 = df1[df1.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df3,x= 'Installs' ,y='App' )
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)
    
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
    
```
